```python
study_hours = int(input("Enter available study hours: "))

subjects = input("Enter subjects separated by commas: ").split(",")

weak_subject = input("Enter your weakest subject: ")
```

```python
print("\n📚 AI Study Plan\n")

for subject in subjects:

    subject = subject.strip()

    if subject.lower() == weak_subject.lower():
        print(f"Spend extra time on: {subject}")
    else:
        print(f"Revise: {subject}")

print(f"\nTotal Study Hours Available: {study_hours}")
time_per_subject = study_hours / len(subjects)

print(f"\nSuggested Study Time Per Subject: {time_per_subject:.1f} hours")
```
